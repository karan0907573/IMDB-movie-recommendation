```python
import pandas as pd
import pickle
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from pathlib import Path

# Download necessary NLTK data
try:
    nltk.data.find('tokenizers/punkt_tab')
except LookupError:
    nltk.download('punkt_tab')
    nltk.download('stopwords')

def preprocess_text(text):
    """
    Cleans the storyline text by removing punctuation, stop words, 
    and unnecessary characters[cite: 24, 25].
    """
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters [cite: 25]
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize the text [cite: 26]
    tokens = word_tokenize(text)
    # Remove stop words [cite: 25]
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [w for w in tokens if w not in stop_words]
    
    return " ".join(filtered_tokens)

# 1. Load the dataset [cite: 18, 21]
df = pd.read_csv('../Data/imdb_movies_2024.csv')

# 2. Data Preprocessing [cite: 22, 76]
# Apply the cleaning function to the Storyline column
df['Cleaned_Storyline'] = df['Storyline'].apply(preprocess_text)
df[['Title', 'Cleaned_Storyline']].to_csv('../Data/imdb_movies_2024_cleaned.csv', index=False)  # Save cleaned data for future use
# 3. Text Representation using TF-IDF [cite: 27, 28]
# Convert storylines into numerical vectors
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df['Cleaned_Storyline'])
with open('../Models/tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(tfidf, f)

with open('../Models/tfidf_matrix.pkl', 'wb') as f:
    pickle.dump(tfidf_matrix, f)

```
