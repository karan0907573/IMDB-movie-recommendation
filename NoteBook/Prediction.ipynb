```python
import pickle
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')

with open("../Models/tfidf_matrix.pkl", "rb") as f:
    tfidf_matrix = pickle.load(f)
with open("../Models/tfidf_vectorizer.pkl", "rb") as f:
    tfidf = pickle.load(f)
```

```python
def preprocess_text(text):
    """
    Cleans the storyline text by removing punctuation, stop words, 
    and unnecessary characters[cite: 24, 25].
    """
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [w for w in tokens if w not in stop_words]
    
    return " ".join(filtered_tokens)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
df=pd.read_csv("../Data/imdb_movies_2024.csv")
```

```python
def get_recommendations(user_input, tfidf_vectorizer, tfidf_matrix, df):
    """
    Calculates Cosine Similarity to find the top 5 most similar movies 
    based on user input[cite: 30, 34].
    """
    # Preprocess the user's input storyline
    cleaned_input = preprocess_text(user_input)
    
    # Transform input into a TF-IDF vector [cite: 78]
    user_tfidf = tfidf_vectorizer.transform([cleaned_input])
    
    # Calculate Cosine Similarity between input and all movies [cite: 31, 80]
    # Similarity score formula: $S_C(A, B) = \frac{A \cdot B}{\|A\| \|B\|}$
    similarity_scores = cosine_similarity(user_tfidf, tfidf_matrix).flatten()
    
    # Get indices of the top 5 most similar movies [cite: 32, 34]
    # Sort in descending order and skip the input itself if it's already in the dataset
    related_indices = similarity_scores.argsort()[:-6:-1]
    
    # Return the top 5 recommended movies [cite: 40, 42]
    return df.iloc[related_indices][['Title', 'Storyline']]

# --- Example Usage ---
# Example user input based on project guidelines [cite: 86]
user_query = "american hero"
recommendations = get_recommendations(user_query, tfidf, tfidf_matrix, df)

print("Top 5 Recommended Movies:")
print(recommendations)
```
